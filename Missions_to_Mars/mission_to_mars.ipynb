```python
# URL of page to be scraped    https://marshemispheres.com/images/full.jpg  /https://marshemispheres.com/cerberus.html
hemi_url = 'https://marshemispheres.com/'
browser.visit(hemi_url)
```

```python
# HTML Object
#html_hemispheres = browser.html
html_hemispheres= browser.html
# Parse HTML with Beautiful Soup
soup3 = BeautifulSoup(html_hemispheres, 'html.parser')
```

```python
# Retreive all items that contain mars hemispheres information
#hemi_url = 'https://marshemispheres.com/'
#find_h=browser.find_by_css('h1[class="site-headline"] img')
items = soup3.find_all('div', class_='item')
#items = browser.find_by_css('div[class="item"]')
# Create empty list for hemisphere urls 
hemisphere_image_urls = []

#html_hemispheres = url
# Store the main_ul 
#hemispheres_main_url = 'https://astrogeology.usgs.gov'

# Loop through the items previously stored
#print(f'my items are: {items}')
for item in items: 
    h_img_url =""
    try:
        # Store title
        title = item.find('h3').text
        #title = browser.find_by_css('h3')
        #print(f'my title is: {title}')
        # Store link that leads to full image website
        img_url_link = item.find('a', class_='itemLink product-item')['href']
        if img_url_link:
            img_url_link = hemi_url + img_url_link
            #print(f'my link is: {img_url_link}')
            #browser.click_link_by_href(img_url_link)
            #splinter loads page
            browser.visit(img_url_link)  # do not need to open?  html_hemi_2= browser.html
            time.sleep(2)

            try:
                #splinter finds the image
                find_image = browser.find_by_css('img[class="wide-image"]')
                #print(f'my image: {find_image}')
                if find_image:
                    #print("inside image object")
                    #find_h=browser.find_by_css('h1[class="site-headline"] img')
                    #for i in find_h:
                    #print(i["src"])
#                     for i in find_image:
#                         #print(i["src"])
#                         h_img_url = i["src"]
                    
                    h_img_url = str(find_image['src'])
                    #print(f'my image url: {image_url}')
                    # Append the retreived information into a list of dictionaries 
                    hemisphere_image_urls.append({"title" : title, "img_url" : h_img_url})
                    #print(hemisphere_image_urls)
                    #img_url = browserfindbytext('img', class_='wide-image')['src']
                    #img_url = html_image_page + soup4.find('img', class_='wide-image')['src']
                
            except Exception as e:
                print(f'hemi error: {e}')
            
    except Exception as e:
        print(e)
        
# Display hemisphere_image_urls
hemisphere_image_urls
```
